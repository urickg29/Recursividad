```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

n = int(input(' Introde un numero entero: '))
print(fibonacci(n))   
```

```python
def sumar_lista(lista):
    if len(lista) == 0:
        return 0
    if  len(lista) == 1:
        return lista[0]
    else:
        return lista[0] + sumar_lista(lista[1:])

numeros = [] #INGRESAR LA LISTA DE NUMEROS AQUI
resultado = sumar_lista(numeros)
print('La suma de la lista es %i' % resultado)
```
